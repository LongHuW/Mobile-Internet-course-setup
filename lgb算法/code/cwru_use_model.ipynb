```python
###使用已保存的模型
import joblib
import numpy as np
import pandas as pd
params = {}
##测试集特征提取后的文件路径
params['test_data_path']=r'E:\文件夹\大三下\移动互联网课设\大作业\408训练\feature\time+fre.csv'
###模型存储路径
params['model_path']=r'E:\文件夹\大三下\移动互联网课设\大作业\time+fre_lgb.model'
###最终识别结果的上传文件路径，可直接上传
params['csv_path']=r'E:\文件夹\大三下\移动互联网课设\大作业\time+fre_lgb.csv'
```

```python
test = np.array(pd.read_csv(params['test_data_path'],keep_default_na=False))##后面这个参数可以解决空值带来的问题
test_x = test[:, :-1]
test_y=test[:,-1]#获取文件标签
all_file_number=int(test_y[len(test_y)-1]-4)#获取文件数量，标签默认从5开始
every_file_number=np.zeros(all_file_number+1)#计数每一个测试文件的特征数量数组，留一个初始0，便于后面循环
predict_index=np.zeros(all_file_number+1)##真正的索引，即上一个数组的依次求和
#获得每个测试文件的特征数量
for file_label in test_y:
    every_file_number[int(file_label)-4]+=1
###获得索引
index_count=0
for i in range(all_file_number+1):
    index_count+=every_file_number[i]
    predict_index[i]=index_count
```

```python
##获得最大概率的预测以及该预测的概率
def get_predict(predict):
    label_kinds=np.zeros(4)
    for label_kind in predict:
        label_kinds[int(label_kind)]+=1
    ##获取四种分类中最多的那个
    label_out=int(np.where(label_kinds==max(label_kinds))[0][0])##加索引是为了避免出现两个预测值概率并列的情况，随便取的第一个，
                                                                ##加int是为了改类型，方便写入csv文件
    #并获得其所占百分比，即他是正确预测的概率
    label_out_rate="%.2f%%" %(label_kinds[label_out]/len(predict)*100)
#     label_out_rate=0
    return label_out,label_out_rate
```

```python
###lgb
my_lgb_model=joblib.load(filename=params['model_path'])
model_predict=my_lgb_model.predict(test_x)
###加头顺便更新该csv文件：
pd.DataFrame([['label','filename']]).to_csv(params['csv_path'],index=False,header=False)
for i in range(all_file_number):
    predict_piece=model_predict[int(predict_index[i]):int(predict_index[i+1])]
    label,label_rate=get_predict(predict_piece)
    print('label:'+str(label)+'rate:'+str(label_rate))
    #写入数据
    a=[]
    a.append(label)
    a.append('TEST'+str(i+1))
    b=[]
    b.append(a)
    pd.DataFrame(b).to_csv(params['csv_path'],mode='a',index=False,header=False)
```
